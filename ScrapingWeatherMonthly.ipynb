```python
## In this code, we are going to perform a web scraping of the wunderground website to obtain the monthly weather data of any city in the world

# import necessary libraries
from selenium import webdriver
import time
import re
import warnings
import pandas as pd
warnings.filterwarnings("ignore")
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')

### URL of the monthly weather data - this is the mother link - later we add the sub URL for the city and month 
url = "https://www.wunderground.com/history/monthly/"
```

```python
## Get soup function
def get_soup(url,q, totaldays):
    url = url + q # Merging the search url and the requested ranking list (q)
    print(url)
    #Initializing Chrome driver
    driver = webdriver.Chrome(chrome_options=options)  
    driver.get(url)  
    # execute script to scroll down the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    # sleep for 10s to allow the driver to read the whole webpage        
    time.sleep(10)  

    rData = []
    
    for row in range(2,totaldays+2):
        for col in range(1,7):
            for col2 in range(1,4):
                iterat = driver.find_elements_by_xpath("/html/body/app-root/app-history/one-column-layout/wu-header/sidenav/mat-sidenav-container/mat-sidenav-content/div/section/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]/table/tbody/tr/td["+str(col)+"]/table/tr["+str(row)+"]/td["+str(col2)+"]")
                for webElement in iterat :
                    rData.append(webElement.text)

    return rData 
```

```python
## Here you can paste the URL of your desired city and the month
### Example of Barcelona in April 2020
#### https://www.wunderground.com/history/monthly/es/el-prat-de-llobregat/LEBL/date/2020-4
q = 'es/el-prat-de-llobregat/LEBL/date/2020-4'
url = url + q # Merging the search url and the requested ranking list (q)
totaldays = 28
Data = get_soup(url,q,totaldays)
```

```python
# Changing the format to float
Data = [float(i) for i in Data]
```

```python
df = pd.DataFrame(columns=['Day', 'T-MAX', 'T-AVG', 'T-MIN','D-MAX', 'D-AVG', 'D-MIN','H-MAX', 'H-AVG', 'H-MIN',
                          'W-MAX', 'W-AVG', 'W-MIN','P-MAX', 'P-AVG', 'P-MIN'])
## Storing the data into a dataframe
for i in range(0,totaldays):
    df = df.append({'Day': Data[i*16], 'T-MAX':Data[i*16 + 1], 'T-AVG':Data[i*16 + 2], 'T-MIN':Data[i*16 + 3],
              'D-MAX':Data[i*16 + 4], 'D-AVG':Data[i*16 + 5], 'D-MIN':Data[i*16 + 6],'H-MAX':Data[i*16 + 7],
              'H-AVG':Data[i*16 + 8], 'H-MIN':Data[i*16 + 9],
              'W-MAX':Data[i*16 + 10], 'W-AVG':Data[i*16 + 11], 'W-MIN':Data[i*16 + 12],
              'P-MAX':Data[i*16 + 13], 'P-AVG':Data[i*16 + 14], 'P-MIN':Data[i*16 + 15]},ignore_index=True)
```

```python
## Changing the day type to Integer
df.astype({'Day': 'int'})
```

```python
## Saving the dataframe into a file
df.to_csv('Barcelona4.csv', index=False)
```
